```python
#craamos graficos. 
import math
import matplotlib.pyplot as plt
import numpy as np
  #generamos los datos   para el grafico 
x=np.array(range(500))*0.1
y=np.zeros(len(x))
for i in range(len(x)):
  y[i]=math.sin(x[i])
   #creamos graficos 
plt.plot(x,y, linestyle)
plt.title('grafiquito')


plt.show()

```
